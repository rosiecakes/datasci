```python
words = ['software:', 'sadly', 'we', 'did', 'adopt', 'from', 'the', 'construction', 'ana \
logy', 'googles', 'stock', 'split', 'means', 'more', 'control', 'for', 'larry', \
 'and', 'sergey', 'ssl', 'software']
```

```python
import re
import operator
```

```python
words = [re.sub('[^a-zA-Z]+', '', word) for word in words]
```

```python
counts = {}

for word in words:
    if word in counts:
        counts[word] += 1
    else:
        counts[word] = 1
```

```python
counts
```

```python
s = sorted(counts.items(), key=operator.itemgetter(1))
```

```python
s.reverse()
```

```python
s[:5]
```

```python
for name, num in counts.items():
    print("{}: {}".format(name, num))
```
